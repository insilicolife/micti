```python
def getTPM(rowCountData, gene_Names=None, index_column=None, ensembol_gene=False, withGene_version=False, specie="human"):

    if(isspmatrix(rowCountData)):
        rowCountData=pa.DataFrame(rowCountData.toarray())
        rowCountData.index=gene_Names
        
    else:
        if index_column is not None:
            rowCountData.index=rowCountData[index_column]
            rowCountData=rowCountData.drop([index_column], axis=1)
            
            
    #read gene annotation from file
    #if(organisum=="hsapiens"):
    #    geneNames=pa.read_csv("https://media.githubusercontent.com/media/insilicolife/micti/master/data/mart_export_stable_genes_human.txt", sep="\t")
    #elif(organisum=="mmusculus"):
    #    geneNames=pa.read_csv("https://media.githubusercontent.com/media/insilicolife/micti/master/data/mouse_mart_export.txt", sep=",", index_col="Gene stable ID")
        
    #else:
    #    print("Please give organisum name")
    #    sys.exit()
    
    if ensembol_gene:
        known_genes=list(map(pyensembl.common.is_valid_ensembl_id,gene_Names))
        rowCountData=rowCountData.iloc[known_genes,:]
        rowCountData.index=np.array(gene_Names)[np.array(known_genes)]
    
        g_clinet=biothings_client.get_client("gene")
        gene_annot=g_clinet.getgenes(list(rowCountData.index), species=specie, fields='symbol,genomic_pos', as_dataframe=True)
        #geneNames=geneNames.groupby(geneNames.index).agg("first")
        #geneNames["Gene stable ID"]=list(geneNames.index)
    
        gene_annot["Gene_len_in_kb"]=(gene_annot["genomic_pos.end"]-gene_annot["genomic_pos.start"])/1000
        
        gene_Names=np.array(gene_Names)[np.array([not i for i in gene_annot["notfound"]==True])]
        rowCountData=rowCountData.iloc[[not i for i in gene_annot["notfound"]==True],:]  
        rowCountData.index=list(gene_Names)
        
        #print(gene_annot)
    else:
        gene_annot=g_clinet.querymany(gene_Names, scopes='symbol', fields='symbol,genomic_pos', species=specie, as_dataframe=True)
        gene_Names=np.array(gene_Names)[np.array([not i for i in gene_annot["notfound"]==True])]
        rowCountData=rowCountData.iloc[[not i for i in gene_annot["notfound"]==True],:]  
        rowCountData.index=list(gene_Names)
        
        gene_annot["Gene_len_in_kb"]=(gene_annot["genomic_pos.end"]-gene_annot["genomic_pos.start"])/1000
        #assert(gene_annot.shape[0]==rowCountData.shape[0])
        
    #print(geneNames.head())
    #get gene length information
    #gene_length=geneNames.loc[list(rowCountData.index),:].dropna()["len_inkb"]
    #count_data=rowCountData.loc[list(gene_length.index),:]

    #count_data_RKB=rowCountData.div(list(gene_length), axis=0)
    count_data_RKB=rowCountData.div(list(gene_annot["Gene_len_in_kb"]), axis=0)
    count_data_TPM=count_data_RKB.div(count_data_RKB.sum(axis=0), axis=1)*10e6
    #print(geneNames.head())
    #count_data_TPM=count_data_TPM[~count_data_TPM.index.duplicated(keep='first')]
    return count_data_TPM.T, list(count_data_TPM.index)
```
